```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

#### Another example from numpy documentation
https://numpy.org/doc/stable/reference/generated/numpy.linspace.html


```python
N = 8
y = np.zeros(N)
x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
fig, ax = plt.subplots()
ax.plot(x1, y, 'o')
ax.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 1])
plt.show()
```

#### Another example from the documentation - Using Explicit Interface

https://matplotlib.org/stable/users/explain/api_interfaces.html#api-interfaces

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1,2)
axs[0].plot([1,2,3], [0, 0.5, 0.2])
axs[1].plot([3,2,1], [0, 0.5, 0.2])
# equivalent to: 
# fig, (ax1, ax2) = plt.subplots(1,2)
# ax1.plot([1,2,3], [0, 0.5, 0.2])
# ax2.plot([3,2,1], [0, 0.5, 0.2])
fig.suptitle('Explicit Interface')
for i in range(2):
    axs[i].set_xlabel('Boo')
plt.show()
```
