### Types of inputs to plotting functions

Matplotlib allows you provide the **data** keyword argument and generate plots passing the strings corresponding to the x and y variables.

```python
np.random.seed(19680801) # seed the random number generator
data = {
    'a': np.arange(50),
    #'c': np.random.randint(0, 50, 50),
    'c': ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h','q', 't', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h','q', 't','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h','q', 't','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h','q', 't','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h','q', 't'],
    'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.rand(50)
data['d'] = np.abs(data['d']) * 100
import pandas as pd
df = pd.DataFrame(data)
df.head()
```

```python

# plot
fig, ax = plt.subplots(figsize=(3, 2.5), layout='constrained') #figsize in inches
ax.scatter('a', 'b', c='c', s='d', data=data) 
#if "c" is a column of the dataframe that contains strings -> ValueError: 'c' argument must be a color, a sequence of colors, or a sequence of numbers
#if "c" is a column of the dataframe that contans numbers -> matplotlib assign automatically the colors to the numbers
# TODO: valibrary plotting: it type_value_colunm is str then our method ok but if int pass to matplotlib directly
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
ax.set_title("Scatter plot")
plt.show()
```

#### Coding styles - A simple line plot

 In general, we suggest using the OO style, particularly for complicated plots, and functions and scripts that are intended to be reused as part of a larger project. However, the pyplot style can be very convenient for quick interactive work.


```python
import matplotlib.pylab as plt
import numpy as np

x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(3, 2.5), layout='constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x_label')
ax.set_ylabel('y_label')
ax.set_title('Simple line plot')
ax.legend()
plt.show()
```

### Making a helper functions

If you need to **make the same plots over and over again with different data sets**, or want to easily wrap Matplotlib methods, use the recommended signature function below, which you would then use twice to populate two subplots.

```python
# GOOD IDEA FOR PERFORMANCE TESTS DEFAULT PLOT
def my_plotter(ax, x_data, y_data, **kwargs):
    # add your personalized plotting
    my_color = 'green'
    ax_out = ax.plot(x_data, y_data, color= my_color, **kwargs)
    return ax_out
```

```python
import matplotlib.pyplot as plt
import numpy as np
data1, data2, data3, data4 = np.random.randn(4, 100) # make 4 random data sets of 100 items each
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(3, 2.5))
my_plotter(ax1, data1, data2, marker = 'x')
my_plotter(ax2, data3, data4, marker = 'o')
ax1.set_xlabel("test")
plt.show()
```

### Colors 

#### Specifyng colors - "CN" color selection

We use a specific style for matplotlib.

A "Cn" color spec, i.e. 'C' followed by a number, which is an index into the default property cycle     

rcParams["axes.prop_cycle"] (default: cycler('color', ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']))) 

The indexing is intended to occur at rendering time, and defaults to black if the cycle does not include color.

https://matplotlib.org/stable/tutorials/colors/colors.html

https://matplotlib.org/stable/gallery/color/color_demo.html




```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
# rcParams["axes.prop_cycle"] (default: cycler('color', 
#                           ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', 
#                            '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']))
x = np.linspace(0, 2 * np.pi, 128)

def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots(figsize=(3,3))
    ax.set_title(f'style: {sty}', color='C0')
    ax.plot(x, np.cos(x), 'C1', label='color1')
    ax.plot(x, np.sin(x), 'C2', label='color2')
    ax.grid(visible=True) # show the cuadricule grid
    ax.legend()

demo('default')
demo('seaborn-v0_8')
```

#### Styling with cycler 

https://matplotlib.org/stable/tutorials/intermediate/color_cycle.html

```python
from cycler import cycler
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi,  4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
yy.shape
# So yy[:, i] will give you the i-th offset sine curve. 
```

```python
mpl.style.use('default') # apply the default style

# let's define the default pror_cycle using matplotlib.pyplot.rc()
default_cycler = (cycler(color=['r', 'g', 'b', 'y']) +
                  cycler(linestyle=['-', '--', ':', '-.']))
plt.rc('lines', linewidth=4)
plt.rc('axes', prop_cycle=default_cycler) # apply the default cycler to all axes

# using the default cycler
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(3, 2.5))
ax0.plot(yy)
ax0.set_title('Set default color to rgby')

# apply a custom cycler
custom_cycler = (cycler(color=['c', 'm', 'y', 'k']) +
                 cycler(lw=[1, 2, 3, 4]))
ax1.set_prop_cycle(custom_cycler)
ax1.plot(yy)
ax1.set_title('Set axes color cycle to cmyk')

# use again the default cycler
ax2.plot(yy)
ax2.set_title('Set default color to rgby')

fig.subplots_adjust(hspace=0.6) # add more space between the plots
plt.show()
```
